```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sn
import scipy.stats as st

sn.set(font_scale=1.4)#for label size

from matplotlib.ticker import MaxNLocator
%matplotlib inline

import random
```

```python
def matches(pile, c):
    m = 0
    for color in ["R", "G", "B", "Y"]:
        if pile.count(color) == c:
            m += 1
    return m
```

```python
hist2 = {}
hist3 = {}

cards = []
for color in ["R", "G", "B", "Y"]:
    for i in range(25):
        cards.append(color)
        
for trial in range(100000):
    random.shuffle(cards)
    
    hand = []
    while not matches(hand, 3) > 0:
        hand.append(cards.pop())
        
    hist2[len(hand)] = hist2.get(len(hand), 0) + 1
    
    while not matches(hand, 4) > 0:
        hand.append(cards.pop())
        
    hist3[len(hand)] = hist3.get(len(hand), 0) + 1
    
    cards += hand
    
for k in hist2:
    hist2[k] /= 100000
for k in hist3:
    hist3[k] /= 100000
```

```python
for i in range(5):

    hist4 = {}

    for trial in range(100000):
        random.shuffle(cards)

        hand = []
        while not matches(hand, 3) > i and not matches(hand, 4) > 0:
            hand.append(cards.pop())

        x = len(hand)
        if matches(hand, 4) == 1:
            x = 4
        if i == 0 and len(hand) == 2:
            x = 4
        hist4[x] = hist4.get(x, 0) + 1

        cards += hand

    for k in hist4:
        hist4[k] /= 100000
        
    mean = 0
    for k in hist4:
        mean += hist4[k] * k
        
    plt.figure()
    plt.bar(list(hist4.keys()), hist4.values())
    plt.title("Mean = " + str(mean))
    plt.ylabel("Probability")
    plt.xlabel("Number of Cards")
```
